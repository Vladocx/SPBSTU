```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("books.csv")

df_2023 = df[df['Год'] == 2023]
df_2024 = df[df['Год'] == 2024]

plt.subplot(2, 1, 1)
plt.plot(df_2023['Месяц'], df_2023['Книги'], marker='D', color='g', label="Прочитано книг")
plt.xlabel("Месяцы")
plt.ylabel("Количество книг")
plt.title("Прочитанные книги 2023")
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.bar(df_2024['Месяц'], df_2024['Страницы'], color = "r")
plt.xlabel("Месяцы")
plt.ylabel("Количество страниц")
plt.title("Прочитанные страницы 2024")
plt.grid(True)

plt.tight_layout()
plt.show()
```
