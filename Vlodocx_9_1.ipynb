```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("books.csv")

df_2023 = df[df[''] == 2023]
df_2024 = df[df[''] == 2024]

plt.subplot(2, 1, 1)
plt.plot(df_2023[''], df_2023[''], marker='D', color='g', label=" ")
plt.xlabel("")
plt.ylabel(" ")
plt.title("  2023")
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.bar(df_2024[''], df_2024[''], color = "r")
plt.xlabel("")
plt.ylabel(" ")
plt.title("  2024")
plt.grid(True)

plt.tight_layout()
plt.show()
```
