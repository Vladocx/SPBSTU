```python
from keybert import KeyBERT
import pandas as pd

def find_keywords(csv_name):
    df = pd.read_csv(csv_name)

    model = KeyBERT('blanchefort/rubert-base-cased-sentiment')

    df["key_words"] = model.extract_keywords(df["normalized_text"])
    return df

find_keywords('cafe_reviews.csv')
```
